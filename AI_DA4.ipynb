```python

Ml = 3		
cl = 3		
mr=0	
cr=0		 
M1 = 0	 
C1 = 0	 
k = 0
print("\nM M M C C C |	 --- | \n")
try:
	while(True):
		while(True):
			print("L->R")

			uM = int(input("M=> "))
			uC = int(input("C=>"))

			if((uM==0)and(uC==0)):
				print("Error")
			elif(((uM+uC) <= 2)and((Ml-uM)>=0)and((cl-uC)>=0)):
				break
			else:
				print("Wrong input re-enter : ")
		Ml = (Ml-uM)
		cl = (cl-uC)
		mr += uM
		cr += uC

		print("\n")
		for i in range(0,Ml):
			print("M ",end="")
		for i in range(0,cl):
			print("C ",end="")
		print("| --> | ",end="")
		for i in range(0,mr):
			print("M ",end="")
		for i in range(0,cr):
			print("C ",end="")
		print("\n")

		k +=1

		if(((cl==3)and (Ml == 1))or((cl==3)and(Ml==2))or((cl==2)and(Ml==1))or((cr==3)and (mr == 1))or((cr==3)and(mr==2))or((cr==2)and(mr==1))):
			print("LOST")

			break

		if((mr+cr) == 6):
			print("WON")
			break
		while(True):
			print("R->L")
			M1 = int(input("M=>"))
			C1 = int(input("C=>"))
			
			if((M1==0)and(C1==0)):
					print("error")
			elif(((M1+C1) <= 2)and((mr-M1)>=0)and((cr-C1)>=0)):
				break
			else:
				print("error")
		Ml += M1
		cl += C1
		mr -= M1
		cr -= C1

		k +=1
		print("\n")
		for i in range(0,Ml):
			print("M ",end="")
		for i in range(0,cl):
			print("C ",end="")
		print("| <-- | ",end="")
		for i in range(0,mr):
			print("M ",end="")
		for i in range(0,cr):
			print("C ",end="")
		print("\n")

	

		if(((cl==3)and (Ml == 1))or((cl==3)and(Ml==2))or((cl==2)and(Ml==1))or((cr==3)and (mr == 1))or((cr==3)and(mr==2))or((cr==2)and(mr==1))):
			print("LOST")
			break
except EOFError as e:
	print("\nEXIT ERROR")
```
